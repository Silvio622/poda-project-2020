## Advantage for Irish farmers
For Irish farmers to produce the cheapest animal feed for milk production is grass. It will be continue for the foreseeable future. The ability to produce grass is the primary competitive advantage over other European dairy farmers. To optimise profitability, producers must maximise the proportion of grazed grass in their cows’ diet. Increasing the proportion of grazed grass in the diet of a dairy cow by 10% can reduce the cost of production by 2.5 cent/litre. The mild Irish climate all year around is ideally to produce quality grass. Important is to match the calving pattern with the grass growing season. The most calves should be born between the 10th of February and 1st of March. Early spring grass is extremely digestible and high in crude protein. To capitalise on the benefits of grazed grass, dairy cows should be turned out to pasture directly after calving if the ground conditions are permitting this. The main objectives of spring grazing management is to increase the proportion of
grazed grass in the diet of the dairy cows. Weather pattern in spring are very important to decide when the best time to start with paddock grazing is. If the soil conditions are too wet On/Off grazing is one strategy to increase the proportion of grazed grass in the cow’s diet. Cows should be grazed 30% in February, 66% by middle of March and the target for 100% should be by beginning middle of April if soil and weather condition permits. From late April onwards grass supply is not restricted.
Further increase of animal performance can be archived by improving the pasture quality. The main objective during the main grazing season is to maximise animal performance from an all grass diet while at the same time maintaining pasture quality[1].

## Irish weather data simulation
In the project I create weather data for the air temperature and the soil temperature with the python “numpy” package. For the air temperature and the soil temperature data simulation I use the “numpy.random.normal” distribution. The weather data’s are based on the Athenry weather station[2]. The “numpy.random.normal” or “bell-shape” distribution has two arguments/ inputs the mean and the standard deviation. The majority of the numerical values around 60% lie centred around the mean. You get the middle part of the “bell”. When the numerical values move farther out on both side of the mean you find less values. The fewer values represent the downward slope on each side of the “bell”.  For the rain fall and the milk yield I use a combination of the “numpy.random.normal” and “numpy.random.uniform” distribution. The “numpy.random.uniform” distribution is used to make the difference between the data of the months a bit smoother. The “numpy.random.uniform” distribution has three arguments the Low, High and Size parameter. In the “Uniform distribution” all the outcomes are likely equal. The data’s will be evenly spread between the Low and High value. But the value of the high variable is not included in the random generated numbers.

```python
import numpy as np
from datetime import date,timedelta
import datetime 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

dictdataset = {} # create a dictionary for the dataset

# Creating dates for the project
listdate=[] # create a list for the dates

startdate = date(2020, 3, 1)   # create a start date variable in format (yyyy,mm,dd)(%y,%m,%d)
enddate = date(2020, 6, 28)   # create a end date variable in format (yyyy,mm,dd)(%y,%m,%d)

deltadate = enddate - startdate       # as timedelta 

for i in range(deltadate.days + 1): # it loops of the amount of days between endate and startdate + 1
    day = startdate + timedelta(days=i) #days = i for the loop index or each loop element
    strday = str(day) # convert day to string
    convdate = datetime.datetime.strptime(strday,("%Y-%m-%d")).strftime("%d/%m/%Y") # convert in different format
    listdate.append(convdate) # put the new coverted date in a list

dictdataset["Date"] = listdate # put the list with dates in a dictionary with the key "Date"
  

#Create random  Air Temperature values
airtemplist = [] # create a list for the air temperature values
airtempcount = 0 # create a variable airtempcount for increasing the date values for the days 

for ele in range(120):
    if airtempcount < 30:
        airtemperature = np.random.normal(7.5,1.5) # mean air temperature for March with a standard deviation of 1.5
        airtemperature =np.round(airtemperature,1) # round random values to one decimal digit 
        airtemplist.append(airtemperature)
        airtempcount = airtempcount + 1
        
    elif airtempcount >= 30 and airtempcount <= 60:
        airtemperature = np.random.normal(8.4,1.5) # mean air temperature for April with a standard deviation of 1.5
        airtemperature =np.round(airtemperature,1) # round random values to one decimal digit 
        airtemplist.append(airtemperature)
        airtempcount = airtempcount + 1

    elif airtempcount >= 60 and airtempcount <= 90:
        airtemperature = np.random.normal(12.4,1.5) # mean air temperature for May with a standard deviation of 1.5
        airtemperature =np.round(airtemperature,1) # round random values to one decimal digit 
        airtemplist.append(airtemperature)
        airtempcount = airtempcount + 1
    
    elif airtempcount >= 90 and airtempcount <= 120:
        airtemperature = np.random.normal(14.4,1.7) # mean air temperature for June with a standard deviation of 1.7
        airtemperature =np.round(airtemperature,1) # round random values to one decimal digit 
        airtemplist.append(airtemperature)
        airtempcount = airtempcount + 1

dictdataset["Air Temperature in 'C"] = airtemplist # put the list with dates in a dictionary with the key 
                                                    # "Air Temperature in 'C"
#Create random Soil Temperature values
soiltemplist = [] # create a list for the soil temperature values
soiltempcount = 0 # create a variable soiltempcount for increasing the date values for the days 

for ele in range(120):
    if soiltempcount < 30:
        soiltemperature = np.random.normal(6.5,1.0) # mean soiltemperature for March with a standard deviation of 1.0
        soiltemperature =np.round(soiltemperature,1) # round random values to one decimal digit 
        soiltemplist.append(soiltemperature)
        soiltempcount = soiltempcount + 1
        
    elif soiltempcount >= 30 and soiltempcount <= 60:
        soiltemperature = np.random.normal(8.2,1.0) # mean soiltemperature for April with a standard deviation of 1.0
        soiltemperature =np.round(soiltemperature,1) # round random values to one decimal digit 
        soiltemplist.append(soiltemperature)
        soiltempcount = soiltempcount + 1

    elif soiltempcount >= 60 and soiltempcount <= 90:
        soiltemperature = np.random.normal(11.3,1.1) # mean soiltemperature for May with a standard deviation of 1.1
        soiltemperature =np.round(soiltemperature,1) # round random values to one decimal digit 
        soiltemplist.append(soiltemperature)
        soiltempcount = soiltempcount + 1
    
    elif soiltempcount >= 90 and soiltempcount <= 120:
        soiltemperature = np.random.normal(13.5,1.1) # mean soiltemperature for June with a standard deviation of 1.1
        soiltemperature =np.round(soiltemperature,1) # round random values to one decimal digit 
        soiltemplist.append(soiltemperature)
        soiltempcount = soiltempcount + 1

dictdataset["Soil Temperature in 'C"] = soiltemplist # put the list with dates in a dictionary with the key 
                                                     # "Soil Temperature in 'C"
#Create random Rain fall values
rainfalllist = [] # create a list for the rain fall values
rainfallcount = 0 # create a variable rainfallcount for increasing the date values for the days 

for ele in range(120):
    if rainfallcount < 25:
        rainfall = np.random.normal(75,1.0) # mean rainfall for March with a standard deviation of 1.0
        rainfall =np.round(rainfall,1) # round random values to one decimal digit 
        rainfalllist.append(rainfall)
        rainfallcount = rainfallcount + 1
    
    elif rainfallcount >= 25 and rainfallcount <= 35:
        rainfall = np.random.uniform(65,72) # mean rainfall for March/April with Low 65 and High 72
        rainfall =np.round(rainfall,1) # round random values to one decimal digit 
        rainfalllist.append(rainfall)
        rainfallcount = rainfallcount + 1

        
    elif rainfallcount >= 35 and rainfallcount <= 60:
        rainfall = np.random.normal(62,1.0) # mean rainfall for April with a standard deviation of 1.0
        rainfall =np.round(rainfall,1) # round random values to one decimal digit 
        rainfalllist.append(rainfall)
        rainfallcount = rainfallcount + 1

    elif rainfallcount >= 60 and rainfallcount <= 70:
        rainfall = np.random.uniform(65,68) # mean rainfall for May with Low 65 and High 68
        rainfall =np.round(rainfall,1) # round random values to one decimal digit 
        rainfalllist.append(rainfall)
        rainfallcount = rainfallcount + 1

    elif rainfallcount >= 70 and rainfallcount <= 90:
        rainfall = np.random.normal(70,1.1) # mean rainfall for May with a standard deviation of 1.1
        rainfall =np.round(rainfall,1) # round random values to one decimal digit 
        rainfalllist.append(rainfall)
        rainfallcount = rainfallcount + 1
    
    elif rainfallcount >= 90 and rainfallcount <= 120:
        rainfall = np.random.normal(70,1.1) # mean rainfall for June with a standard deviation of 1.1
        rainfall =np.round(rainfall,1) # round random values to one decimal digit 
        rainfalllist.append(rainfall)
        rainfallcount = rainfallcount + 1

dictdataset["Rain fall in mm"] = rainfalllist # put the list with dates in a dictionary with the key 
                                              # "Rain Fall in mm"
#Create random values for daily milk yield per cow 
milkpercowlist = [] # create a list for the rain fall values
milkpercowcount = 0 # create a variable milkpercowcount for increasing the date values for the days 

for ele in range(120):
    if milkpercowcount < 25:
        milkpercow = np.random.normal(12,0.5) # mean milkpercow for March with a standard deviation of 0.5
        milkpercow =np.round(milkpercow,1) # round random values to one decimal digit 
        milkpercowlist.append(milkpercow)
        milkpercowcount = milkpercowcount + 1

    elif milkpercowcount >= 25 and milkpercowcount <= 35:
        milkpercow = np.random.uniform(12,15) # mean milkpercow for March?April with with Low 12 and High 15
        milkpercow =np.round(milkpercow,1) # round random values to one decimal digit 
        milkpercowlist.append(milkpercow)
        milkpercowcount = milkpercowcount + 1
        
    elif milkpercowcount >= 35 and milkpercowcount <= 60:
        milkpercow = np.random.normal(16,1.0) # mean milkpercow for April with a standard deviation of 1.0
        milkpercow =np.round(milkpercow,1) # round random values to one decimal digit 
        milkpercowlist.append(milkpercow)
        milkpercowcount = milkpercowcount + 1

    elif milkpercowcount >= 60 and milkpercowcount <= 90:
        milkpercow = np.random.normal(18,1.1) # mean milkpercow for May with a standard deviation of 1.1
        milkpercow =np.round(milkpercow,1) # round random values to one decimal digit 
        milkpercowlist.append(milkpercow)
        milkpercowcount = milkpercowcount + 1
    
    elif milkpercowcount >= 90 and milkpercowcount <= 120:
        milkpercow = np.random.normal(20,1.3) # mean milkpercow for June with a standard deviation of 1.3
        milkpercow =np.round(milkpercow,1) # round random values to one decimal digit 
        milkpercowlist.append(milkpercow)
        milkpercowcount = milkpercowcount + 1

dictdataset["Milk per cow in litres"] = milkpercowlist # put the list with dates in a dictionary with the key 
                                                       # "Milk per cow in litres"

# Create a Dataset
df = pd.DataFrame.from_dict(dictdataset)
print(df)

```

```python
# Plot the Dataframe
fig = plt.figure(figsize=(14,10))
plt.style.use('ggplot')

ax1 = fig.add_subplot(2,2,1)
sns.scatterplot(data=df, x = "Date", y = " Air Temperature in 'C", ax = ax1, color='r');
ax1.set_title(" Air Temperature in 'C")
ax1.set(xlabel="Date")
ax1.xaxis.set_major_locator(plt.MaxNLocator(5))


#plt.subplot(2,2,2)
ax2 = fig.add_subplot(2,2,2)
sns.scatterplot(data=df, x = "Date", y = "Soil Temperature in 'C", ax = ax2, color='g');
ax2.set_title("Soil Temperature in 'C")
ax2.set(xlabel="Date")
ax2.xaxis.set_major_locator(plt.MaxNLocator(5))

ax3 = fig.add_subplot(2,2,3)
sns.scatterplot(data=df, x = "Date", y = "Rain fall in mm", ax = ax3, color='b');
ax3.set_title("Rain fall in mm")
ax3.set(xlabel="Date")
ax3.xaxis.set_major_locator(plt.MaxNLocator(5))

ax4 = fig.add_subplot(2,2,4)
sns.scatterplot(data=df, x = "Date", y = "Milk per cow in litres", ax = ax4, color = 'y');
ax4.set_title("Milk per cow in litres")
ax4.set(xlabel="Date")
ax4.xaxis.set_major_locator(plt.MaxNLocator(5))
```

### Correlation between Air and Soil Temperature
The grass starts growing by a soil temperature of above 4'C. The soil temperature is an average 1'C lower than the air temperature. The soil temperature is moving almost accordingly with the air temperature by 1'C by looking at the regression lines below for the two temperatures. In February should be enough grass grow that cows can be grazed 30%, by middle of March enough grow for 66% and by beginning middle of April enough for 100% grassing. From late April onwards grass supply is not restricted anymore. 
***

```python
meanairtemp = df.mean(axis=0)["Air Temperature in 'C"]
meansoiltemp = df.mean(axis=0)["Soil Temperature in 'C"]
print('Air Temperature:{}'.format(meanairtemp))
print('Soil Temperature:{}'.format(meansoiltemp))
```

```python
# Insert a regression line for the air temperture and the soil temperature
coefftemp = np.polyfit(df.index,airtemplist,1)# use dataframe index for x values and air

coeffsoil = np.polyfit(df.index,soiltemplist,1)# or use index


fig1 = plt.figure(figsize=(14,6))
plt.style.use('ggplot')

# Correlation between Air and Soil Temperature plot
ax5 = fig1.add_subplot(1,2,1)
plt.scatter(x= "Date",y = "Air Temperature in 'C", c='r',data=df)
plt.plot(df.index,coefftemp[0] * df.index + coefftemp[1],'-', c='r')
plt.scatter(x= "Date",y = "Soil Temperature in 'C", c='g',data=df)
plt.plot(df.index,coeffsoil[0] * df.index + coeffsoil[1],'-', c='g')
plt.legend()
plt.title("Correlation between Air and Soil Temperature")
plt.xlabel("Date")
plt.ylabel("Temperature in 'C")
ax5.xaxis.set_major_locator(plt.MaxNLocator(5))


# Increase of Milk Production
ax6 = fig1.add_subplot(1,2,2)
plt.scatter(x= "Date",y = "Milk per cow in litres", c='y',data=df)
plt.legend()
plt.title("Increase of Milk Production")
plt.xlabel("Date")
plt.ylabel("Milk per cow in litres")
ax6.xaxis.set_major_locator(plt.MaxNLocator(5))
plt.axvline(x=df.index[15], color='grey', linestyle=':') # create vertical line
plt.axvline(x=df.index[45], color='grey', linestyle=':') # create vertical line
plt.text(x = df.index[1], y = 22, s = "30%")
plt.text(x = df.index[0], y = 21.6, s = "grazing")
plt.text(x = df.index[25], y = 22, s = "66%")
plt.text(x = df.index[24], y = 21.6, s = "grazing")
plt.text(x = df.index[61], y = 22, s = "100%")
plt.text(x = df.index[60], y = 21.6, s = "grazing")

plt.show()

```
