```python
# Create a Temperature profile for Athenry Ireland from April until July

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
tempApril=[]

# mean (mu) temperature 8'C  and standard deviation (sigma) 1.5'C samplesize 30 days
mu, sigma, samplesize = 8, 1.5, 30 
tempApril = np.random.normal(mu, sigma, samplesize)
tempApril

plt.subplot(1,1,1) 
sns.distplot(tempApril);
```
